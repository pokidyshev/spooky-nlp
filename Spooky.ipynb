```python
sns.countplot(x=train.author);
```

```python
eap = train[train.author=='EAP'].text
hpl = train[train.author=='HPL'].text
mws = train[train.author=='MWS'].text
```

```python
sns.set_style('whitegrid', {'axes.grid' : False})
```

```python
WordCloud(stopwords=STOPWORDS, min_font_size=8, width=1000, height=400)\
    .generate(' '.join(eap))\
    .to_image()
```

```python
WordCloud(stopwords=STOPWORDS, min_font_size=8, width=1000, height=400, colormap='cubehelix_r')\
    .generate(' '.join(hpl))\
    .to_image()
```

```python
WordCloud(stopwords=STOPWORDS, min_font_size=8, width=1000, height=400, colormap='hot')\
    .generate(' '.join(mws))\
    .to_image()
```
